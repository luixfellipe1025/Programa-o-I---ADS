```python

"""
Luiz Felipe Paniz Rodrigues - 24-02-2024
Atividade desplugada
Criando uma flor com pixels no Python
"""
import numpy as np
import matplotlib.pyplot as plt

# Criando a matriz 15x9 (altura x largura) para representar a imagem
imagem = np.zeros((15, 9, 3))  # Inicialmente, todos os pixels são pretos (0,0,0)

# Definindo cores RGB para cada parte da flor
amarelo = [1, 0.5, 0.5]   # Pétalas da Flor
verde = [0, 1, 0]     # Caule
marrom = [0.5, 0.25, 0]  # Terra
verde = [0, 1 ,0]  # Miolo da Flor

# Preenchendo as regiões correspondentes com as cores adequadas
imagem[5:8, 3:6] = amarelo  # Pétalas da Flor
imagem[8:13, 4:5] = verde   # Caule
imagem[12:13, 3:6] = marrom  # Terra
imagem[6:7, 4:5] = verde   # Miolo da Flor

# Definindo o fundo da imagem como branco
imagem[np.where((imagem == [0, 0, 0]).all(axis=2))] = [1, 1, 1]

# Exibindo a imagem
plt.imshow(imagem)

# Configurações adicionais para melhor visualização
plt.grid(color='gray', linestyle='-', linewidth=1)
plt.xticks(np.arange(-0.5, 9, 1), [])
plt.yticks(np.arange(-0.5, 15, 1), [])
plt.tick_params(
    axis='both',
    which='both',
    bottom=False,
    top=False,
    left=False,
    right=False,
    labelbottom=False,
    labelleft=False)
plt.xlim(-0.5, 8.5)
plt.ylim(14.5, -0.5)

plt.show()
```
